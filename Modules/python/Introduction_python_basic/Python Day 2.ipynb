```python
class Car:
    """Parent class having basic methods of car"""
    def color(self):
        """Return color of the car"""
        return "Red"

    def wheels(self):
        """Return the number of wheels in car"""
        return 4

class Mahindra(Car):
    """Child class extending the basic methods of car"""
    def brand(self):
        """Brand of the car"""
        return "Mahindra"

class Hyundai(Car):
    """Child class extending basic methods of car"""
    def brand(self):
        """Brand of the car"""
        return "Hyundai"

    def color(self):
        """Override the default color behavior""" 
        return "white"

car=Car()
print(car.color())

mahindra=Mahindra()
print(mahindra.brand())
print(mahindra.color())

hyundai=Hyundai()
print(hyundai.color())
 
```

# Method

Method is block of code within the class which can perform predefine set of actions and provide better modularity and reusability.

```python
class MyClass:
    variable = "blah"

    def method1(self):
        """simple finction"""
        print("This is a message inside the class.")
        
    def method2(self, input_string):
        """Passing input variables"""
        print(f"input string is: {input_string}")
    
    def method3(self, input_string=None):
        """Passing input variable with default value as None"""
        print(f"no input" if input_string is None else input_string)
        
    def method4(self, a, b=4):
        """Passing input without value and with default value"""
        print(f"a+b = {a+b}")
    
    def method5(self, a=None, b=None):
        """Both variables with None value"""
        return a + b if a and b else None
    
    def method6(self, *args):  
        """uncertain number of variables"""
        for i in args:
            print(f"variable arg: {i}")
    
    def method7(self, **kwargs):
        """ It is used when we want to pass a variable number of keyword arguments"""
        for k, v in kwargs.items():
            print(f"variable keyword-arg: key:{k} and value:{v}") 
```

```python
obj1 = MyClass()
# obj1.method1()
obj1.method2("hello world")
obj1.method3()
obj1.method4(a= 3,b= 5)
obj1.method4(b=3,a=5)
obj1.method4(3,b = 6)
obj1.method5(3,b = 7)
obj1.method5(3,b = None)
obj1.method6(10,"Two",30,50)
obj1.method7(**{"a": "hello world", "b" : 55})
```

```python
class Circle:
    pi = 3.14159  # Class attribute

    def __init__(self, radius):
        self.radius = radius  # Instance attribute

    # Instance method
    def area(self):
        return self.pi * self.radius ** 2

    # Instance method
    def circumference(self):
        return 2 * self.pi * self.radius

    # Class method
    @classmethod
    def from_diameter(cls, diameter):
        radius = diameter / 2
        return cls(radius) # Circle(5)

    # Static method
    @staticmethod
    def is_positive(number):
        return number > 0

# # Creating an instance of the Circle class
# circle1 = Circle(5)
# print(f"Area of circle1: {circle1.area()}")  # Output: Area of circle1: 78.53975
# print(f"Circumference of circle1: {circle1.circumference()}")  # Output: Circumference of circle1: 31.4159

# Using the class method
circle2 = Circle.from_diameter(10)
print(f"Area of circle2: {circle2.area()}")  # Output: Area of circle2: 78.53975

# Using the static method
print(Circle.is_positive(10))  # Output: True
print(Circle.is_positive(-5))  # Output: False
```

## Constructor in Python

```python
class Employee:
    """Basic employee class"""
    def __init__(self, emp_name, emp_id):
        """Constructor/initializer for employee"""
        self.emp_name = emp_name
        self.emp_id = emp_id

    def _id(self):
        """Return employee id"""
        return self.emp_id

    def name(self):
        """Return employee name"""
        return self.emp_name

employee = Employee(1, "Rishabh")
print(employee._id(), employee.name())

employee2 = Employee(2, "Yogesh")
print(employee2._id(), employee2.name())
```

# Pass By Reference vs Value

### Pass By Value:
 A copy of the object is passed to the function instead of the actual object.

```python
def square(x):
    x = x * x
    print("Inside square function: ", x)

num = 5
square(num)
print("Outside square function: ", num)
```

The value of num remains the same outside the function even though x was modified inside the function. This is because the value of num was passed to the square function, and a copy of the value was created so that the function could modify it.

### Pass By Reference:

Function gets an implicit reference to the argument, rather than a copy of its value. i.e. function can modify the argument.

This is generally true for mutable object types such as: List, Dict, Set, etc.


In Python, everything is an object, and variables are just references or pointers to those objects. Therefore, when you pass an argument to a function, its reference is passed, and not the actual value.

```python
import random
def by_reference(my_dict):
    # loop through all the keys in the dictionary
    for key in my_dict:
        # print out each key and its current value
        print(f"variable keyword-arg: key:{key} and value: {my_dict.get(key)}")
        # update the value of the current key to a new value
        num = random.randrange(1, 101)
        my_dict[key] = num
    # return the updated dictionary
    return my_dict
```

```python
# create a dictionary
student_scores = {
    "Yogesh": 85,
    "Rishabh": 92,
    "Prashant": 99
}

# print out the original dictionary
print("Original Dictionary:", student_scores)

# pass the dictionary to the by_reference function
new_scores = by_reference(student_scores)

# print out the new dictionary returned by the function
print("New Dictionary:", new_scores)

# print out the original dictionary after the function call
print("Original Dictionary after Function Call:", student_scores)
```

## Multiple Inehritance in Detail with MRO

In python (unlike java), a class can have two or more parent classes, this is known as multiple inheritance

Multiple inheritance creates confusion, as to which parent class method will be executed, if two or more parent classes have a method with the same name

<span>**<span style="color:red">NOTE:</span> Method  Resolution Order(MRO) is used to resolve which parent method to call**</span>

```python
class A:
    def foo(self):
        print("A.foo()")

class B(A):
    def foo(self):
        print("B.foo()")

class C(A):
    def foo(self):
        print("C.foo()")

class D(B, C):
    pass

# Checking the MRO
print(D.__mro__)
print(D)

# Creating an instance of D
d = D()
print(d)
d.foo()
```

```python
class A:
    def foo(self):
        print("A.foo()")

class B(A):
    pass

class C(A):
    def foo(self):
        print("C.foo()")

class D(C, B):
    pass

# Creating an instance of D
print(D.__mro__)
d = D()
d.foo()
```

<span>**<span style="color:red">Question:</span> Does python MRO follow Depth First Search?**</span>

<span>MRO follows something that is called **C3 Linerization** which is nither DFS or BFS, it is a combination of both</span>

<span>**In Summary:** C3 Linerization is depth-first until classes are encountered that will share a parent, and then breadth-first over those</span>

### List Comprehensions:

comprehension means inclusion.

List comprehension is way to define and create list. It is a more humanly way of creating a list.

It will make your code more concise (concise means giving a lot of information clearly and in a few words).

```python
list9 = [x**2 for x in range(10)]
print(list9)
list9 = [x**2 for x in range(1,10)]
print(list9)
list9 = [x**2 for x in range(1,10,2)]
print(list9)
```

<span style="color:red">**Thumb Rule:**</span> List comprehension always start with square bracket and end with square bracket. And the output of any list comprehension is a list.

```python
# Traditional way of creating list of square of even number
list10 = []
for element in range(1,10):
    if element%2==0:
        list10.append(element**2)
print(list10)
```

```python
# List comprehension way of getting the square of even number
list11 = [element**2 for element in range(1,10) if element%2==0]
print(list11)
```

## Dictionary Iteration:

We can loop over the dictionary keys, values, and the key-value pairs

```python
dict1 = {"one": 1,
         "two": 2,
         "three": 3}

print(dict1)

# Iterating over keys
for key in dict1.keys():
    print(key)
# Iterating over values
for value in dict1.values():
    print(value)
# Iterating over key-value pairs
for key, value in dict1.items():
    print(key, value)
```

#### Limitation on dict size manipulation while looping

It is not possible to manipulate the length of a dictionary while looping

```python
dict1 = {"one": 1,
         "two": 2,
         "three": 3}
dict1["four"] = 4
print(dict1)
for key, value in dict1.items():
    if value == 1:
        dict1["four"] = 11
print(dict1)
```
